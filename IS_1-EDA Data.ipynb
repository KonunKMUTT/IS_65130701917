```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm

# Assuming df is your DataFrame

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
plt.suptitle('Age', fontsize=25, fontweight='bold', color='navy')
fig.tight_layout(pad=2.0)

# Plotting histogram with normal curve
sns.histplot(data=df, x='age', kde=True, ax=axes[0])  # Add kde=True for kernel density estimate

# Fitting a normal distribution to the data
mu, std = norm.fit(df['age'])
xmin, xmax = axes[0].get_xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)

# Plotting the normal curve
axes[0].plot(x, p, 'k', linewidth=2)

# Plotting boxplot
sns.boxplot(data=df, x='age', y='class', ax=axes[1])

plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming 'gender' column in df
gender_counts = df['gender'].value_counts()


# Plotting pie chart
plt.figure(figsize=(10, 5))
plt.pie(gender_counts, labels = ['Male', 'Female'], autopct='%1.1f%%', startangle=90, colors=['blue', 'pink'])
plt.suptitle('Gender Distribution', fontsize=15, fontweight='bold', color='navy')


plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'gender' and 'class' columns in df
plt.figure(figsize=(10, 5))
ax = sns.countplot(x='gender', hue='class', data=df)

# Calculate percentages and annotate the plot
total = len(df['gender'])
for p in ax.patches:
    height = p.get_height()
    plt.text(p.get_x() + p.get_width() / 2., height + 3,
             f'{height/total:.1%}', ha='center')

# Set legend labels
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, ['Negative', 'Positive'], title='Class')

# Set explicit tick positions and labels for the x-axis
ax.set_xticks([0, 1])
ax.set_xticklabels(['Female', 'Male'])

plt.suptitle('Gender Distribution by Class', fontsize=15, fontweight='bold', color='navy')
plt.show()
```

```python
import numpy as np
from scipy import stats  # Import the stats module

# Assuming 'troponin' and 'kcm' columns are present in df
z_scores = stats.zscore(df[['troponin', 'kcm']])
abs_z_scores = np.abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
df = df[filtered_entries]
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
plt.suptitle('Creatine Kinase-MB', fontsize=25, fontweight='bold', color='navy')
fig.tight_layout(pad=2.0)

# Plotting histogram with normal curve
sns.histplot(data=df, x='kcm', kde=True, ax=axes[0])  # Add kde=True for kernel density estimate

# Fitting a normal distribution to the data
mu, std = norm.fit(df['kcm'])
xmin, xmax = axes[0].get_xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)

# Plotting the normal curve
axes[0].plot(x, p, 'k', linewidth=2)

# Plotting boxplot
sns.boxplot(data=df, x='kcm', y='class', ax=axes[1])

plt.show()
```
